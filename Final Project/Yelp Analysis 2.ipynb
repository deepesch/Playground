```python
def cleanText(text):
    """
    removes punctuation, stopwords and returns lowercase text in a list of single words
    """
    text = text.lower()    
    
    from bs4 import BeautifulSoup
    text = BeautifulSoup(text).get_text()
    
    from nltk.tokenize import RegexpTokenizer
    tokenizer = RegexpTokenizer(r'\w+')
    text = tokenizer.tokenize(text)
    
    from nltk.corpus import stopwords
    clean = [word for word in text if word not in stopwords.words('english')]
    
    return clean
```

```python
data['cleantext'] = data['text'].apply(cleanText)
```

```python
 X_train, X_test, y_train, y_test = train_test_split(data['cleantext'], data.rating) #Train-Test Split
```


The Yelp dataset contains lots of features for each restaurant. However, most of them are categorical variables, which could potentially be considered bad for making predictions from these variables.

Our approach to tackle the "categorical variables" problem, is to construct continuous variables from the existing dataset. 

As experimented the data with all sorts of the models and got insignificant output, I started to wonder: If none of these techniques performs good, what does this tell us? Does it infer the dataset itself is non-informative in terms of predicting stars? 



Then naturally, we question: what makes the predicting power of reviews so powerful?
In order to answer this question, we use Topic Modelling to analyze the reviews. From the output of Topic Modelling, we can see that the reviews are all about food, not enviroment, or waiter/waitress, etc. And we do not have the food taste/quality information in the original dataset!

We can further infer that the most important thing that customers care about is not the restaurant's environment, not the decoration, not how friendly the waiter/waitress is, not the location, but the--food!!!
